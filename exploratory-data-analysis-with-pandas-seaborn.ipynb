# *Exploratory Data Analysis*

# Gender

```python
ax = sns.countplot(x = "Gender", data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

> Most of the orders were occured by female customers(7832)

```python
sales_based_on_gender = df.groupby(["Gender"],  as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False)

ax1 = sns.barplot(x = "Gender", y = "Amount", data = sales_based_on_gender)

# for bars in ax1.containers:
#     ax1.bar_label(bars)
```

> In amount of purchasing, the female customers are higher than male customers

```python
ax2 =  sns.countplot(x = "Age Group", hue = "Gender", data = df)

for bars in ax2.containers:
    ax2.bar_label(bars)
```

```python
sales_based_on_age = df.groupby(["Age Group"],  as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False)

ax1 = sns.barplot(x = "Age Group", y = "Amount" , data = sales_based_on_age)
```

> Buyes from age 26 to 35 purchased most amount and Female Buyer Greater than Male Buyer

# State

```python
sales_based_on_state = df.groupby(["State"],  as_index = False)["Orders"].sum().sort_values(by="Orders", ascending = False).head(10)

sns.set(rc = {"figure.figsize": (28, 12)})
ax3 = sns.barplot(x = "State", y = "Orders", data = sales_based_on_state)

for bars in ax3.containers:
    ax3.bar_label(bars)
```

```python
sales_based_on_state = df.groupby(["State"],  as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False).head(10)

sns.set(rc = {"figure.figsize": (28, 12)})
ax3 = sns.barplot(x = "State", y = "Amount", data = sales_based_on_state)

for bars in ax3.containers:
    ax3.bar_label(bars)
```

> Most order and amount were placed by Uttar Pradesh, Maharashtra and Karnataka

# Marital_Status

```python
ax4 = sns.countplot(x = "Marital_Status", data = df)
sns.set(rc = {"figure.figsize": (7, 5)})

for bars in ax4.containers:
    ax4.bar_label(bars)
```

```python
sales_based_on_marital_status = df.groupby(["Marital_Status","Gender"],  as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False).head(10)
sns.set(rc = {"figure.figsize": (7, 5)})
ax4 = sns.barplot(x = "Marital_Status", y = "Amount", hue="Gender", data = sales_based_on_marital_status)

# for bars in ax3.containers:
#     ax3.bar_label(bars)
```

> Married Women ordered and purchased most

# Occupation

```python
ax5 = sns.countplot(x = "Occupation", data = df)
sns.set(rc = {"figure.figsize": (24, 9)})
for bars in ax5.containers:
    ax5.bar_label(bars)
```

```python
sales_based_on_occupation = df.groupby(["Occupation"],  as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False).head(10)

sns.set(rc = {"figure.figsize": (28, 12)})
ax3 = sns.barplot(x = "Occupation", y = "Amount", data = sales_based_on_occupation)

for bars in ax3.containers:
    ax3.bar_label(bars)
```

Purchasing power is high in IT Sector, Healthcare and Aviation

```python
ax6 = sns.countplot(x = "Product_Category", data = df)
sns.set(rc = {"figure.figsize": (24, 7)})
for bars in ax5.containers:
    ax6.bar_label(bars)
```

```python
sales_based_on_occupation = df.groupby(["Product_Category"],  as_index = False)["Amount"].sum().sort_values(by="Amount", ascending = False).head(10)

sns.set(rc = {"figure.figsize": (28, 12)})
ax5 = sns.barplot(x = "Product_Category", y = "Amount", data = sales_based_on_occupation)

for bars in ax5.containers:
    ax5.bar_label(bars)
```

> In Food category most amount was occured

# Conclution
